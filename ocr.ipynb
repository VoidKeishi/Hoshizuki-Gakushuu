```python
from PIL import Image
import pytesseract
import os
from matplotlib import pyplot as plt

from matplotlib.font_manager import FontProperties
```

```python
# Print the text next to the image
def draw_ocr_comparison_figs(img, text):
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
    # Display the image on the left side
    ax1.imshow(img)
    ax1.axis('off')  # Turn off axis labels
    ax1.set_title('Original Image')

    # Display the OCR text on the right side with a Japanese font
    font_prop = FontProperties(fname=r'fonts\\GenJyuuGothicMonospaceBold.ttf')  # Replace with the path to your Japanese font
    ax2.text(0.1, 0.5, text, fontsize=20, verticalalignment='center', fontproperties=font_prop)
    ax2.axis('off')  # Turn off axis labels
    ax2.set_title('OCR Text')

    # Adjust layout for better spacing
    plt.tight_layout()

    # Show the plot
    plt.show()
```

```python
# Set the path to the Tesseract executable (change this to the path on your system)
pytesseract.pytesseract.tesseract_cmd = "D:\\GitHub\\Hoshizuki-Gakushuu\\TesseractOCR\\tesseract.exe"
# Open an image using Pillow
image_paths = [os.path.join('images', image_path) for image_path in os.listdir('images') if image_path.endswith('.png')]

for image_path in image_paths:
    img = Image.open(image_path)
    # Use pytesseract to do OCR on the image
    text = pytesseract.image_to_string(img, lang='Japanese')
    draw_ocr_comparison_figs(img, text)
```
